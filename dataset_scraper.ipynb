```python
import os
i=0
os.mkdir('Covid-19')
fldr='Covid-19/'
j=0
#print(len(df))
while i<len(df):
    if 'COVID-19' in df.iloc[i]['finding']:
        #print(j)
        #print(df.iloc[i]['finding'])
        fl= df.iloc[i]['folder']+'/'+df.iloc[i]['filename']
        fl_to=fldr+df.iloc[i]['filename']
        instruction='cp '+fl+" "+fl_to
        os.system(instruction)
        j+=1
    i+=1
```
